```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# Assuming data is a pandas DataFrame

# Drop the 'Private' column (or any other identifier column)
data_without_private = data.drop('Private', axis=1)

# Identify categorical columns
categorical_columns = data_without_private.select_dtypes(include=['object']).columns

# Use OneHotEncoder to convert categorical columns to numeric
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_categorical_data = encoder.fit_transform(data_without_private[categorical_columns])

# Create a DataFrame for the encoded categorical data
encoded_categorical_df = pd.DataFrame(encoded_categorical_data, columns=encoder.get_feature_names_out(categorical_columns))

# Drop the original categorical columns from the data
data_numeric = data_without_private.drop(categorical_columns, axis=1)

# Concatenate the numeric data with the encoded categorical data
final_data = pd.concat([data_numeric.reset_index(drop=True), encoded_categorical_df.reset_index(drop=True)], axis=1)

# Optional: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(final_data)

# Initialize the KMeans model with a specified number of clusters
n_clusters = 3  # Choose an appropriate number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the model to the data
kmeans.fit(scaled_data)

# Optional: Get the cluster labels
labels = kmeans.labels_

# Optional: Add the cluster labels back to the original DataFrame
data['Cluster'] = labels

# Optional: Display the first few rows of the DataFrame with the cluster labels
print(data.head())
```
